```python
# Task 2: Mark the success/failed launches for each site on the map
import folium
from folium.features import DivIcon

# Cr√©er la carte centr√©e sur la NASA Johnson Space Center
nasa_coordinate = [29.559684, -95.083097]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Ajouter les marqueurs pour chaque site de lancement unique
for _, launch_site in launch_sites_df.iterrows():
    folium.Circle(
        location=[launch_site['Lat'], launch_site['Long']],
        radius=1000,
        color='blue',
        fill=True,
        fill_opacity=0.7,
        popup=launch_site['Launch Site']
    ).add_to(site_map)
    
    folium.map.Marker(
        [launch_site['Lat'], launch_site['Long']],
        icon=DivIcon(
            icon_size=(150,36),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12pt; color: blue;"><b>{launch_site["Launch Site"]}</b></div>',
        )
    ).add_to(site_map)

# Ajouter les lancements r√©ussis et √©chou√©s
for _, launch in spacex_df.iterrows():
    if launch['class'] == 1:
        color = 'green'
        label = 'Success'
    else:
        color = 'red'
        label = 'Failed'
    
    folium.CircleMarker(
        location=[launch['Lat'], launch['Long']],
        radius=5,
        color=color,
        fill=True,
        fill_opacity=0.7,
        popup=f"{launch['Launch Site']} - {label}"
    ).add_to(site_map)

# Afficher la carte
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
import numpy as np

# Fonction pour calculer la distance en kilom√®tres entre deux points g√©ographiques
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # rayon de la Terre en km
    phi1 = np.radians(lat1)
    phi2 = np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)
    
    a = np.sin(delta_phi/2.0)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(delta_lambda/2.0)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    
    distance = R * c
    return distance

# Exemple: calculer la distance entre chaque lancement et chaque site
launch_site = launch_sites_df.iloc[0]  # Premier site
launch_lat = launch_site['Lat']
launch_lon = launch_site['Long']

# Ajouter une colonne 'Distance_from_site' pour chaque lancement
spacex_df['Distance_from_site'] = spacex_df.apply(
    lambda row: haversine(launch_lat, launch_lon, row['Lat'], row['Long']),
    axis=1
)

# Afficher les 5 premi√®res distances
spacex_df[['Launch Site', 'Distance_from_site']].head()
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
